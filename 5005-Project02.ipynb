```python
import csv

csvfile=open("homeprices.csv")
csvfile2=open("SB.csv")
csvreader=csv.reader(csvfile)
csvreader2=csv.reader(csvfile2)
homePrices=list(csvreader)
SBrklyn=list(csvreader2)

import pandas as pd
import matplotlib.pyplot as plt


NYNeighborhoods=[]
for i in homePrices[1:]:
    if i[4]=='NY' and i[5] == 'NY' and i[6]=='New York' and i[8] == 'Kings County' :
        NYNeighborhoods.append(i)


df = pd.read_csv('SB.csv')

meta_cols = ['RegionName', 'RegionType', 'StateName', 'State', 'City', 'CountyName']
date_cols = [c for c in df.columns if c not in meta_cols]

for _, row in df.iterrows():
    plt.plot(date_cols, row[date_cols].values, label=row['RegionName'])

plt.title('NY Neighborhood Home Prices')
plt.xlabel('Date')
plt.ylabel('ZHVI ($)')
plt.xticks(date_cols[::24], rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```
